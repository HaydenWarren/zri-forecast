```python
import pandas as pd
import numpy as np
import plotly.express as px
```

**City Level Unemployment Rate**: Not seasonally adjusted yet

```python
#San Antonio Unemployment Rate monthly 10yr
SA_ur = pd.read_csv(r'C:\Users\robla\Desktop\nycdsa\Capstone\data_dump\San_Antonio_Unemployment_Rate.csv')
#Houston Unemployment Rate monthly 10yr
HOU_ur = pd.read_csv(r'C:\Users\robla\Desktop\nycdsa\Capstone\data_dump\Houston_Unemployment_Rate.csv')
#El Paso Unemployment Rate monthly 10yr
EP_ur = pd.read_csv(r'C:\Users\robla\Desktop\nycdsa\Capstone\data_dump\El_Paso_Unemployment_Rate.csv')
#Austin Unemployment Rate monthly 10yr
AU_ur = pd.read_csv(r'C:\Users\robla\Desktop\nycdsa\Capstone\data_dump\Austin_Unemployment_Rate.csv')
#Dallas-Fort Worth Unemployment Rate monthly 10yr
DAFW_ur = pd.read_csv(r'C:\Users\robla\Desktop\nycdsa\Capstone\data_dump\Dallas_FW_Unemployment_Rate.csv')
```

```python
list(SA_ur['DATE']) == list(HOU_ur['DATE']) == list(EP_ur['DATE']) == list(AU_ur['DATE']) == list(DAFW_ur['DATE'])
```

```python
dfs = [SA_ur,EP_ur,AU_ur,DAFW_ur]
for i in dfs:
    print(len(i))
```

```python
UR = HOU_ur
for df in dfs:
    UR = pd.merge(UR,df,how = 'left',on = 'DATE')   
    
UR.columns = ['Time','houston_ur',
               'san_antonio_ur',
               'el_paso_ur',
               'austin_ur',
               'dallas_fw_ur']

UR.loc[:,'Time'] = UR['Time'].apply(lambda x: x[:-3])
UR.loc[:,'Time'] = UR['Time'].apply(lambda x: x[x.find('-')+1:] + '-' + x[:x.find('-')])
```

```python
fig = px.line(data_frame = UR, x = 'Time',y = list(UR.columns)[1:6],
              title ='City Unemployment Rates (Not Seasonally Adjusted)',
              width = 1000)
fig.show()
```

### Adjusting for seasonality with percentage change

```python
cols = list(UR.columns)[1:]

for col in cols:
    
    UR.loc[:,'delta_' + col] = UR[col].diff(1)
```

```python
fig = px.line(data_frame = UR, x = 'Time',y = list(UR.columns)[6:],
              title ='Differenced (Period = 1) City Unemployment Rates',
              width = 1000)
fig.show()
```

**Monthly State Level Retail Gas Prices (Regular Formulation) TX**

```python
TX_gas = pd.read_csv\
(r'C:\Users\robla\Desktop\nycdsa\Capstone\data_dump\Texas_Regular_All_Formulations_Retail_Gasoline_Prices.csv')

TX_gas.columns = ['Time','monthly_avg_gas_price']

TX_gas.loc[:,'year'] = TX_gas['Time'].apply(lambda x: '20'+x[x.find('-')+1:])

TX_gas.loc[:,'Month'] = TX_gas['Time'].apply(lambda x: x[:x.find('-')])

TX_gas = TX_gas.reset_index().sort_values(by = 'index', ascending = False)

TX_gas = TX_gas.groupby(by = 'year').apply(lambda g: g.sort_values(by = 'index', ascending = False)).droplevel(-1)

TX_gas = TX_gas.drop('year', axis = 1).reset_index()

ordered_months = list(TX_gas['Month'])[8:20] #reworking month format
imp = dict(zip(ordered_months,np.arange(1,13)))
for k,v in imp.items():
    if v in [1,2,3,4,5,6,7,8,9]:
        imp[k] = '0' + str(v) 
    else:
        imp[k] = str(v) 

TX_gas.loc[:,'Month'] = TX_gas['Month'].apply(lambda x: imp[x])

TX_gas.loc[:,'Time'] = TX_gas['Month']+'-'+TX_gas['year']

TX_gas = TX_gas.drop(['year','Month'], axis = 1)

TX_gas.loc[:,'delta_monthly_avg_gas_price'] = TX_gas['monthly_avg_gas_price'].diff(1)

#TX_gas
```

```python
fig = px.line(data_frame = TX_gas, x = 'Time',y = 'monthly_avg_gas_price',
              title ='Monthly State Level Retail Gas Prices (Regular Formulation) TX',
              width = 1000)
fig.show()
```

```python
fig = px.line(data_frame = TX_gas, x = 'Time',y = 'delta_monthly_avg_gas_price',
              title ='Differenced (Period = 1) Monthly State Level Retail Gas Prices (Regular Formulation) TX',
              width = 1000)
fig.show()
```
