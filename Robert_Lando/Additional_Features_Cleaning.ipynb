**Monthly State Level Retail Gas Prices (Regular Formulation) TX**

```python
TX_gas = pd.read_csv\
(r'C:\Users\robla\Desktop\nycdsa\Capstone\data_dump\Texas_Regular_All_Formulations_Retail_Gasoline_Prices.csv')

TX_gas.columns = ['Time','monthly_avg_gas_price']

TX_gas.loc[:,'year'] = TX_gas['Time'].apply(lambda x: '20'+x[x.find('-')+1:])

TX_gas.loc[:,'Month'] = TX_gas['Time'].apply(lambda x: x[:x.find('-')])

TX_gas = TX_gas.reset_index().sort_values(by = 'index', ascending = False)

TX_gas = TX_gas.groupby(by = 'year').apply(lambda g: g.sort_values(by = 'index', ascending = False)).droplevel(-1)

TX_gas = TX_gas.drop('year', axis = 1).reset_index()


ordered_months = list(TX_gas['Month'])[7:19] #reworking month format
imp = dict(zip(ordered_months,np.arange(1,13)))
for k,v in imp.items():
    if v in [1,2,3,4,5,6,7,8,9]:
        imp[k] = '0' + str(v) 
    else:
        imp[k] = str(v) 

TX_gas.loc[:,'Month'] = TX_gas['Month'].apply(lambda x: imp[x])

TX_gas.loc[:,'Time'] = TX_gas['year']+'-'+TX_gas['Month']

TX_gas.loc[:,'Time'] = TX_gas['Time'].apply(lambda x: pd.to_datetime(x,format = '%Y-%m'))

TX_gas = TX_gas.drop(['year','Month'], axis = 1)

TX_gas.loc[:,'delta_monthly_avg_gas_price'] = TX_gas['monthly_avg_gas_price'].diff(1)

TX_gas.loc[:,'State'] = ['TX']*TX_gas.shape[0]

#TX_gas
```

```python
city_zips.loc[:,'State'] = ['TX']*city_zips.shape[0]
#city_zips
```

```python
TX_gas.shape
```

```python
city_zips.shape
```

```python
monthly_avg_gas_prices = pd.merge(TX_gas,city_zips, how = 'left',on = 'State')
monthly_avg_gas_prices.rename(columns = {'zip':'zip_code'},inplace = True)
monthly_avg_gas_prices = monthly_avg_gas_prices.drop('index', axis = 1)
monthly_avg_gas_prices.head()
```
